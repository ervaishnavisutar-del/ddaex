```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("medicalinsurance.csv")
```

```python
df
```

```python
df["age_group"]=pd.cut(
    df["age"],
    bins=[0,25,40,70],
    labels=["Young","Adult","Senior"]
)
df
```

```python
df["smoker_binary"]=df["sex"].map({"no":0,"yes":1})
df["sex_binary"]=df["sex"].map({"female":0,"male":1})
df
```

```python
df["bmi_category"]=pd.cut(
    df["bmi"],
    bins=[0,18,25,45],
    labels=["underweight","normal","overweight"]
)
df
```

```python
region_encoded=pd.get_dummies(df["region"])
region_encoded
```

```python

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df["age_scaled"]=scaler.fit_transform(df[["age"]])
df["bmi_scaled"]=scaler.fit_transform(df[["bmi"]])
df["children_scaled"]=scaler.fit_transform(df[["children"]])
df

```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df["age_std"]=scaler.fit_transform(df[["age"]])
df["bmi_std"]=scaler.fit_transform(df[["bmi"]])
df["children_std"]=scaler.fit_transform(df[["children"]])
df

```

```python
df.to_csv("medicalinsurance_feature_enginnering.csv",index=False)
```
