```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("dirty-employee.csv")
```

```python
df.head()
```

```python
print("Shape:",df.shape)
```

```python
df.info()
```

```python
df=df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Unnamed: 10'],inplace=True)
```

```python
df['Full_Name'].fillna('Unknown',inplace=True)
```

```python
df['Age'].describe()
```

```python
df['Joining_Date']=pd.to_datetime(df['Joining_Date'],errors='coerce')
```

```python
df['Joining_Date']=df['Joining_Date'].dt.strftime("%Y-%m-%d")
```

```python
df['Salary']=df['Salary'].replace('$','')
```

```python
df['Salary']=df['Salary'].replace(',','')
```

```python
df['Gender']=df['Gender'].replace({
    "M":"Male",
    "F":"Female",
    "male":"Male",
    "female":"Female"
})
```

```python
df['Country']=df['Country'].str.title()
```

```python
df['Phone']=df['Phone'].str.replace(r'[^0-9]',regex=True)
```

```python
df[(df['Age'])>=1&(df['Age']<=100)]
```

```python
df.to_csv('dirty-employee_cleaned.csv',index=False)
```
