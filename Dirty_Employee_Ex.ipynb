```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("dirty-employee.csv")
```

```python
print(df.head())
```

```python
print("Shape:",df.shape)
```

```python
df.info()
```

```python
df=df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Unnamed: 10'],inplace=True)
```

```python
df['Age'].fillna(df['Age'].median(),inplace=True)
```

```python
df['Full_Name'].fillna('Unknown',inplace=True)
```

```python
df['Phone'].fillna('Not Available',inplace=True)
```

```python
df['Marks'].fillna(df['Marks'].median(),inplace=True)
```

```python
df['Full_Name']=df['Full_Name'].str.strip()
```

```python
df['Gender']=df['Gender'].str.title().str.strip()
```

```python
df['Joining_Date']=pd.to_datetime(df['Joining_Date'],errors='coerce')
```

```python
df=df[(df['Age']>=18)&(df['Age']<=60)]
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Salary']=df['Salary'].astype(str).str.replace(',','',regex=False)
```

```python
df['Phone']=df['Phone'].astype(str).str.replace(r'\D','',regex=True)
```

```python
print("\nNull Values After Cleaning:")
print(df.isnull().sum())
```

```python
print('\nCleaned Dataset:')
print(df.head())
```

```python
print('\nDataset Shape:')
print(df.shape)
```

```python
print('\nDataset Information:')
print(df.info())
```

```python
df.to_csv('dirty-employee_cleaned.csv',index=True)
```
