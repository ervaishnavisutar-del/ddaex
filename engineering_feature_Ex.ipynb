```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    "Name":["Alex","Bob","Charlie","David","Ethon"],
    "Age":[22,26,35,40,33],
    "Salary":[22000,35000,45000,160000,40000],
    "City":["Mumbai","Pune","Mumbai","Miraj","miraj"]
}
df=pd.DataFrame(data)
df

```

```python
df["Age.Group"]=pd.cut(
    df["Age"],
    bins=[0,25,40,100],
    labels=["Young","Adult","Senior"]
)
df
```

```python
#Add Salary Category Feature
def Salary_Categories(Salary):
  if Salary<=35000:
      return"Low"
  elif Salary<=75000:
      return"Medium"
  else:
      return"High"
df ["Salary_Category"]= df["Salary"].apply(Salary_Categories)
df
```

```python
# One-Hot Encoding
City_encoded=pd.get_dummies(df["City"])
City_encoded
```

```python
#Combing Data
df["Combined"]=df["Salary"]/df["Age"]
df["Combined"]
```

```python
from sklearn.preprocessing import MinMaxScaler
Scalar=MinMaxScaler()
df["Age_Scaled"]=Scalar.fit_transform(df[["Age"]])
df["Salary_Scaled"]=Scalar.fit_transform(df[["Salary"]])
df
```

```python
from sklearn.preprocessing import StandardScaler
Scalar=StandardScaler()
df["Age_Std"]=Scalar.fit_transform(df[["Age"]])
df["Salary_Std"]=Scalar.fit_transform(df[["Salary"]])
df
```
